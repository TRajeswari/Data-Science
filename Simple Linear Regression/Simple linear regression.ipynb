```python
import numpy as np
age = np.array([[20],[26],[28],[34],[38],[40]])
age
```

```python
weight = np.array([60,66,78,74,80,76])
weight
```

```python
import matplotlib.pyplot as plt
plt.scatter(age,weight)
plt.xlabel("Age")
plt.ylabel("Weight")
plt.show()
```

```python
x = age
y = weight
```

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
```

```python
# model fitting
LR.fit(x,y)
# Bo
print("Intercept value: ",LR.intercept_)
# B1
print("coeffcient value: ",LR.coef_)
```

```python
LR.predict(np.array([[30]]))
```

```python
Y_pred = LR.predict(age)
Y_pred
```

```python
weight
```

```python
import matplotlib.pyplot as plt
plt.scatter(age,weight)
plt.scatter(age,Y_pred, color='red')
plt.plot(age,Y_pred, color='black')
plt.xlabel("Age")
plt.ylabel("Weight")
plt.show()
```

```python
y
```

```python
Y_pred
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y,Y_pred)
print("Mean squared error:", mse.round(2))
```

```python
import numpy as np
RMSE = np.sqrt(mse)
print("Root Mean squared error:", RMSE.round(2))
```
